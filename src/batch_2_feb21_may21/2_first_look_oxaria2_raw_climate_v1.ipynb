```python
# Imports
# ---------
import sys
import pandas as pd
import numpy as np
import feather
import os
import gc
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pandas.plotting import register_matplotlib_converters
import matplotlib.ticker as ticker
from matplotlib.dates import DateFormatter
import matplotlib as mpl
import qgrid

register_matplotlib_converters()

# File locations
# ----------------
the_gases = './oxaria/data/raw/2oxaria/json/gap_filling/batch_2_feb21_may21/'
pngs = './oxaria/data/pngs/'
```

```python
# Load the df
# -------------
oxaria2_climate = pd.read_feather(
    the_gases+'oxaria2_climate_q12021_gf.ftr').set_index(['tag', 'rec']).sort_index()
```

```python
# Temp & RH ts plot
# ------------------
plt.style.use('ggplot')
myFmt = mdates.DateFormatter('%b')
start_date = dt.datetime(2021, 1, 1)
end_date = dt.datetime(2021, 6, 1)

fig, axes = plt.subplots(nrows=5, ncols=2, sharex=False,
                         sharey=False, figsize=(15, 20))
axes_list = [item for sublist in axes for item in sublist]

fig.suptitle('Raw OxAria1 temperature & RH outputs, January 2021 onward',
             fontsize=11.5, y=0.91)

for tag, dat in oxaria2_climate.groupby('tag'):
    ax = axes_list.pop(0)
    dat.reset_index(level=0, inplace=True)
    dat.plot(y='val.tmp', c='tab:red', linewidth=0.5, marker='o', ms=0.1, ls='', label='Temp.', ax=ax, legend=True,
             zorder=1)
    dat.plot(y='val.hmd', c='tab:green', linewidth=0.5, marker='o', ms=0.1, ls='', label='RH', ax=ax, legend=True,
             zorder=1)
    ax.set_title(tag, fontsize=10)
    ax.tick_params(
        axis='x',
        which='major',
        labelrotation=0,
        labelsize=8
    )
    ax.tick_params(axis='y', labelsize=9)
    ax.tick_params(axis='x', which='minor', length=0)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(28))
    ax.xaxis.set_major_formatter(myFmt)
    plt.setp(ax.xaxis.get_majorticklabels(), ha='center')
    ax.set_xlim(start_date, end_date)
    ax.set_ylabel('T & RH (deg C)', fontsize=8)
    ax.set_xlabel('')
    plt.subplots_adjust(wspace=0.14, hspace=0.3)
    ax.legend(markerscale=40, frameon=True, framealpha=0.99, loc='upper right')

plt.savefig(pngs+'oxaria2_trh_raw_sensor_q12021_ts.png')
plt.show()

for ax in axes_list:
    ax.remove()
```
